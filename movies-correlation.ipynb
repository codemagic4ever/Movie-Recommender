```python
import pandas

#naming the columns for a better understanding
#ratings_features = ["user_id", "item_id", "rating", "timestamp"]

#created a ratings dataframe which has 4 columns and 100837 rows
ratings_df = pandas.read_csv("ratings.csv")
                                
ratings_df
```

```python
#index means we are calling the rows and at 0 means the first row
#dropping the first row
ratings_df = ratings_df.drop(ratings_df.index[0])

ratings_df
```

```python
ratings_df.info()
```

```python
#changing the dataframe type to float to allow numeric calculations
ratings_df = ratings_df.astype("float")

ratings_df.info()
```

```python
#all the values now have a decimal since we converted the type to a float
ratings_df
```

```python
#created a movies df which has 3 columns and 9742 rows
movies_df = pandas.read_csv("movies.csv")

movies_df
```

```python
movies_df.info()
```

```python
#going to remove genres column

movie_titles_df = movies_df[["movieId", "title"]]

movie_titles_df
```

```python
movie_titles_df.info()
```

```python
#converted the movieId to float from int type
#went from int to str to float
movie_titles_df["movieId"] = movie_titles_df["movieId"].astype(str).astype(float)

movie_titles_df.info()
```

```python
#combined the rating and the movie dataframes into one by using merge on the common column name movieId
merged_df = pandas.merge(ratings_df, movie_titles_df, on = "movieId")

merged_df
```

```python
#getting the number of times each movieid is referenced
merged_df.groupby("movieId")["rating"].count().sort_values(ascending=False)
```

```python
#making a pivot table 
crosstab = merged_df.pivot_table(values = "rating",
                      index = "userId",
                      columns = "title",
                      fill_value = 0)

crosstab
```

```python
#transposing the crosstab and calling it X
X = crosstab.T

X
```

```python
#to reduce dimensionality of the data

#it performs Singular Value Decomposition (SVD) but only keeps a limited number of components instead of fully factorizing the matrix.
#reducing dimensionality means keeping only the most important patterns while ignoring noise and redundant features.
from sklearn.decomposition import TruncatedSVD
#elbow point of the graph tells you how many components are good or actually add meaningful value
NUMBER_OF_COMPONENTS = 12
#the top 12 values with the most variance are kept which retains the most amount of valuable information
singular_value_decomposition = TruncatedSVD(n_components = NUMBER_OF_COMPONENTS,
             random_state = 1)
#the SVD is fit onto X
matrix = singular_value_decomposition.fit_transform(X)

matrix
```

```python
import numpy

correlation_matrix = numpy.corrcoef(matrix)

correlation_matrix
```

```python
movie_titles = crosstab.columns

movie_titles
```

```python
movies_list = list(movie_titles)

movies_list
```

```python
example_movie_index = movies_list.index("Batman: Year One (2011)")

example_movie_index
```

```python
example_correlations = correlation_matrix[example_movie_index]

example_correlations
```

```python
MAXIMUM_CORRELATION = 1.0

MINIMUM_CORRELATION = 0.7


list(movie_titles[(example_correlations < MAXIMUM_CORRELATION) & (example_correlations > MINIMUM_CORRELATION)])
```
